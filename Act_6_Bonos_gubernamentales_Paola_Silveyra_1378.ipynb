```python
# ============================================
# PROYECTO UIII - IA CON if, elif, else y while
# Tema: Bonos Gubernamentales
# Alumno: Paola Silveyra
# Grupo: 2 H
# ============================================

print("🏦 SIMULADOR DE INTELIGENCIA ARTIFICIAL 🏦")
print("Sistema experto en Bonos Gubernamentales\n")

# Variables
continuar = "si"

# Ciclo principal
while continuar == "si":

    print("\n========== MENÚ PRINCIPAL ==========")

    # Pregunta 1
    print("\n1. ¿Cuánto dinero deseas invertir?")
    inversion = float(input("Ingresa la cantidad en pesos: "))

    # Pregunta 2
    print("\n2. ¿Qué plazo prefieres?")
    print("a) Corto plazo")
    print("b) Mediano plazo")
    print("c) Largo plazo")
    plazo = input("Selecciona una opción (a/b/c): ")

    # Pregunta 3
    print("\n3. ¿Qué nivel de riesgo aceptas?")
    print("a) Bajo")
    print("b) Medio")
    print("c) Alto")
    riesgo = input("Selecciona una opción (a/b/c): ")

    # Pregunta 4
    print("\n4. ¿Cuál es tu objetivo?")
    print("a) Ahorrar")
    print("b) Generar rendimientos")
    print("c) Protegerme de la inflación")
    objetivo = input("Selecciona una opción (a/b/c): ")

    print("\n🤖 ANALIZANDO RESPUESTAS...")

    # Inteligencia Artificial simulada
    if plazo == "a" and riesgo == "a":
        print("\n✅ RECOMENDACIÓN IA:")
        print("Invertir en CETES")
        print("✔ Son seguros")
        print("✔ Ideales para corto plazo")
        print("✔ Bajo riesgo")

    elif plazo == "b" and riesgo == "b":
        print("\n✅ RECOMENDACIÓN IA:")
        print("Invertir en BONOS M")
        print("✔ Buen rendimiento")
        print("✔ Mediano plazo")
        print("✔ Riesgo moderado")

    elif plazo == "c" and riesgo == "c":
        print("\n✅ RECOMENDACIÓN IA:")
        print("Invertir en UDIBONOS")
        print("✔ Protegen contra inflación")
        print("✔ Largo plazo")
        print("✔ Mayor rendimiento")

    elif objetivo == "c":
        print("\n✅ RECOMENDACIÓN IA:")
        print("Invertir en UDIBONOS")
        print("✔ Protegen tu dinero")
        print("✔ Ayudan contra la inflación")

    elif inversion < 1000:
        print("\n✅ RECOMENDACIÓN IA:")
        print("Comenzar con CETES")
        print("✔ Puedes invertir desde poco dinero")

    else:
        print("\n✅ RECOMENDACIÓN IA:")
        print("Invertir en BONDES")
        print("✔ Opción flexible")
        print("✔ Buena para inversionistas principiantes")

    # Otra consulta
    continuar = input("\n¿Deseas realizar otra consulta? (si/no): ")

print("\n👋 Gracias por usar el simulador de IA")
print("Fin del programa")
print("Programa realizado por Paola Silveyra NL = 50")
```
